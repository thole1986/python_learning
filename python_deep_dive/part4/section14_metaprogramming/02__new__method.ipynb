```python
class Person:
    def __new__(cls, name):
        print(f"Person: instantiating {cls.__name__}...")
        instance = object.__new__(cls)
        return instance
    
    def __init__(self, name):
        print("Person: initializing instance...")
        self.name = name
```

```python
p = Person('Tho')
```

```python
p.__dict__
```

```python
class Student(Person):
    def __new__(cls, name, major):
        print(f"Student: instantiating {cls.__name__}...")
        instance = super().__new__(cls)
        return instance
    
    def __init__(self, name, major):
        print(f"Student: instantiating instance...")
        super().__init__(name)
        self.major = major
```

```python
s = Student('Tho', 'Major')
```

```python
class Square:
    def __new__(cls, w, l):
        cls.area = lambda self: self.w * self.l
        instance = super().__new__(cls)
        return instance
    
    def __init__(self, w, l):
        self.w = w
        self.l = l
```

```python
s = Square(3, 4)
```

```python
s.area()
```

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name
    
    def __eq__(self, value):
        return isinstance(value, Person) and self.name == value.name
    
    def __hash__(self):
        return hash(self.name)
    
    def __rerp(self):
        return f"Person(name='{self.name}')"
    
```

```python
p1 = Person('Eric')
```

```python
hash(p1)
```

```python
p2 = Person('Eric')
p1 == p2
```

```python
hash(p1), hash(p2)
```

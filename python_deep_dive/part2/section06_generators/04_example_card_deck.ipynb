````python
from collections import namedtuple

Card = namedtuple('Card', 'rank suit')
SUITS = ('Spades', 'Hearts', 'Dimonds', 'Clubs')
RANKS = tuple(range(2, 11)) + tuple('JQKA')
````

````python
print(RANKS)
````

```
    suit_index = card_index // len(RANKS)  -> Chia lấy phần nguyên
    
    i // len(RANKS) là chia lấy phần nguyên, để xác định “chất” của lá bài.

    Mỗi chất có 13 lá, nên:

    i = 0–12 → i // 13 = 0 → SUITS[0] → 'Spades'

    i = 13–25 → i // 13 = 1 → SUITS[1] → 'Hearts'

    i = 26–38 → i // 13 = 2 → SUITS[2] → 'Diamonds'

    i = 39–51 → i // 13 = 3 → SUITS[3] → 'Clubs'
    rank_index = card_index % len(RANKS)
```

````python
print(len(SUITS))
print(len(RANKS))
print(range(len(SUITS) * len(RANKS)))
````

````python
def card_gen():
    for i in range(len(SUITS) * len(RANKS)):
        suit = SUITS[i // len(RANKS)]
        rank = RANKS[i % len(RANKS)]
        card = Card(rank, suit)
        yield card
````

````python
for card in card_gen():
    print(card)
````

````python
class CardDeck:
    SUITS = ('Spades', 'Hearts', 'Dimonds', 'Clubs')
    RANKS = tuple(range(2, 11)) + tuple('JQKA')

    def __iter__(self):
        return CardDeck.card_gen()
    
    def __reversed__(self):
        return CardDeck.reversed_card_gen()
    
    @staticmethod
    def card_gen():
        for suit in CardDeck.SUITS:
            for rank in CardDeck.RANKS:
                yield Card(rank, suit)
    
    @staticmethod
    def reversed_card_gen():
        for suit in reversed(CardDeck.SUITS):
            for rank in reversed(CardDeck.RANKS):
                yield Card(rank, suit)
````

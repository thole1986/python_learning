```python
import math
```

```python
class FactIter:
    def __init__(self, n):
        self.n = n
        self.i = 0

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        else:
            result = math.factorial(self.i)
            self.i += 1
            return result
```

```python
fact_iter = FactIter(5)
```

```python
list(fact_iter)
```

```python
list(fact_iter)
```

```python
next(fact_iter)
```
